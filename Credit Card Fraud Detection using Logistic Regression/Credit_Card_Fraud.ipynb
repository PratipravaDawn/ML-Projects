##Getting Data

```python
import pandas as pd
data = pd.read_csv("creditcard.csv")
data.head()
```

```python
#Checking the no of datas
data.shape
```

```python
#Checking the data type for each value
data.info()
```

```python
#Checking if null values are present
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
#Checking no of values present
#Converting to integer
data = data.astype({'Class':int})
data.Class.value_counts()
```

```python
#Separating the Legit and Fraud Transactions
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
print(legit.shape, fraud.shape)
```

Legit ---> 0 </br> Fraud ---> 1

##Under-Sampling
Undersampling to make the legit and fraud data equal

```python
legit_sample = legit.sample(n=205)
legit_sample.shape
```

```python
data_concate = pd.concat([legit_sample, fraud], axis=0)
data_concate.shape
```

```python
x = data_concate.drop(['Class', 'Time'], axis=1)
y = data_concate['Class']
print(x, y)
```

##Splitting the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
print(x.shape, x_train.shape, x_test.shape)
```

##Model: Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
predict_y = model.predict(x_test)
```

##Checking the Accuracy Score

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, predict_y)
print(f"The accuracy is {acc :.2f}")
```
